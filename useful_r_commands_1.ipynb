Section 1:
Version control facilitates reproducible analysis. Researchers should strive to use the same version of a software package for data analysis throughout the entirety of a project. This is because there could be changes between versions of a software package that may influence how an analysis is done. The same principle holds when working with programming languages such as R. Another consideration is that some R packages may only work for an earlier but not the current version of R. Thus, there is a need to keep more than one R version installed on our machines and the ability to quickly check what R version we are working with is imporant. Most of us are bench biologists so using the same version of a software or programming language for data analysis throughout a project is similar to using the same reagent or data collection instrument throughout a lab experiment.

Section 2:
Important for those who are just beginning to work with R is to have an understanding of data types and data structures in R. 

```r
## section 2a - exploring R data types: integers, double precision, character
#### for more on integers and double precision see http://uc-r.github.io/integer_double/
#### note that R does not have single precision data type
#### on the surface, it looks like the integer 5 is assigned to variable_1
#### but using the typeof command, we see that variable_1 is actually double precision
#### essentially, all R platforms are required to work with values conforming to IEEE 754 standards (double precision)
#### with double precision, the range is between 2e-308 to 2e+308 (?double in R to read more)
#### see https://en.wikipedia.org/wiki/Double-precision_floating-point_format
#### see https://youtu.be/3RTbnzHtf7Q
variable_1 <- 5
variable_1 
typeof(variable_1)

#### double precision can accomodate NaN
variable_2 <- NaN
variable_2

#### to change variable_1 to an integer, we can use the as.integer command
#### integers have a range of about +/-2*10^9 (?integer in R)
#### so here, we see that double precision offer
variable_3 <- as.integer(variable_1)
variable_3
typeof(variable_3)

#### we can also append L at the end to indicate integer
variable_4 <- 6L
variable_4
typeof(variable_4)

#### to change variable_3 from integer to double precision we can use as.numeric or as.double
variable_5 <- as.numeric(variable_3)
variable_5
typeof(variable_5)

#### variable_6 is still an integer
variable_6 <- 2000000050L
variable_6
typeof(variable_6)
#### variable_7 is double precision
variable_7 <- 3000000000L
variable_7
typeof(variable_7)

#### complex numbers
variable_8 <- 3+1i
variable_8
typeof(variable_8)
#### logical
variable_9 <- TRUE
variable_9
typeof(variable_9)

#### we can also work with characters in R and R has functionalities to manipulate characters
#### for instance, gsub allows us to replace one character with another
#### in this example, we use gsub to find R in variable_10 and replace it with Python
variable_10 <- "this is a R class"
typeof(variable_10)
variable_11 <- gsub("R", "Python", variable_10)
variable_11
```

```r
## Section 2b - vectors in R
#### a vector is an array of elements with one dimension
#### to create a vector in R, use c() and input vector elements in the parentheses
variable_12 <- c(5, 10, 8, 4, 6, 9)
variable_12
typeof(variable_12)
#### retrieve the number of elements in variable_12 using the length command
length(variable_12)
#### retrieve 2nd element in variable_12
variable_12[2]
#### retrieve the 1st and 3rd elements in variable_12
variable_12[c(1,3)]
#### retrieve all elements of variable_12 except for the 2nd
variable_12[-2]
#### changes the 3rd element of variable_12 from 8 to 21
variable_12[3] <- 21 
variable_12
#### using strings as vector indices
variable_13 <- c("a"=3, "b"=12, "c"=4) 
variable_13["a"]
```

```r
## Section 2c - matrices in R
#### a matrix is an m row by n column array of similar data
#### thus, a matrix has two dimensions while a vector has one
#### to reference an element in a matrix we use matrix_name[m,n], where m is the row number, and n is the column number
#### create a 3 row by 3 column matrix and store as variable_14
variable_14 <- matrix(nrow=3,ncol=3)
#### create 3 vectors, variable_15, variable_16, and variable_17, which will be stored as rows in the matrix corresponding to variable_14
variable_15 <- c(3,6,2)
variable_16 <- c(4,8,10)
variable_17 <- c(2,5,7)
variable_14[1,] <- variable_15
variable_14[2,] <- variable_16
variable_14[3,] <- variable_17
#### use det to get the determinant to the matrix stored in variable_14
det(variable_14)
#### use eigen to obtain the eigen values and eigen vectors
#### for those doing RNA seq analysis, the concept of eigen values and eigen vectors will be important for principle components analysis (PCA)
eigen(variable_14)
#### retrieve the element in row 2 and column 2 of the matrix stored in variable_14
variable_14[2,2]
#### retrieve column 3 of the matrix stored in variable_14
variable_14[,3]
#### retrieve row 3 of the matrix stored in variable_14
variable_14[3,]
#### multiply matrix in variable 14 with vector stored in variable 18
#### note that we have to use %*%, otherwise using * we will end up with another 3 by 3 matrix
#### for a quick tutorial on matrix algebra see https://youtu.be/MG7t6SWBnwA 
variable_18 <- c(2,3,6)
variable_19 <- variable_14%*%variable_18
variable_19
#### we can assign column and row names to a matrix
colnames(variable_14) <- c("a", "b", "c")
rownames(variable_14) <- c("d", "e", "f")
variable_14
#### note that we can not reference the columns in a matrix using column name
variable_14$a
```

```r
## Section 2d - data frames in R
#### a data frame is used to store data tables and contains multiple data types in multiple columns 
#### think of a data frame as an Excel sheet
#### we can convert the matrix stored in variable_14 to a data frame using data.frame
variable_21 <- data.frame(variable_14)
variable_21$gene_names <- c("TP53", "NF-KB", "BRCA1") ## creates a new column called gene_names
variable_21 <- variable_21[,c(4,1:3)] ## arranges column order so that gene_names shows up first
colnames(variable_21)[2:4] <- c("sample_a", "sample_b", "sample_c")
variable_21
#### notice that we can now reference the columns using column name
variable_21$sample_a
```

```r
## Section 2e - lists
#### as a heads up, lists will be important when you start working with packages like edgeR for RNA seq
#### we can create a list with random things
list_1 <- list(3, "a", "dogs", "cats")
list_1
typeof(list_1)
list_1[[1]]
typeof(list_1[[1]])
#### we can also create a list that stores data frames
variable_22 <- data.frame(gene_name=c("ADA", "AMPD2", "ADSS"),counts=c(100,250,300))
variable_23 <- data.frame(gene_name=c("EPAS1","IRP1","EPO"),counts=c(200,150,175))
list_2 <- list(variable_22, variable_23)
list_2[[2]]
```

```r
## Section 6a - getting data frame dimensions
#### nrow will retrieve the number of rows in a data frame
#### ncol will retrieve the number of columns in a data frame
nrow(time_series_covid19_confirmed_us)
ncol(time_series_covid19_confirmed_us)
```

```r
## Section 6b - subsetting and column name change
#### the us covid confirmed cases data frame has 663 columns (so very large)
#### let's subset the confirmed cases data frame before working with it
#### grab only county, state, and case numbers from 10/30/2021 to 11/3/2021
time_series_covid19_confirmed_us_1 <- time_series_covid19_confirmed_us[,c(6,7,659:663)]
## change column header Admin2 to County
colnames(time_series_covid19_confirmed_us_1)[1] <- "County" 
head(time_series_covid19_confirmed_us_1)
```

```r
## Section 6c - filtering
#### getting the case counts from 10/31/2021 to 11/3/2021 for Maryland only using grep
#### note that in R, we can use grep to search for patterns as we can do with unix command line
#### $ is used to reference a particular column in R data frames
#### we can use grep to filter because it will return the row indices for Province_State which equal Maryland
time_series_covid19_confirmed_maryland_1 <- time_series_covid19_confirmed_us_1[grep("Maryland", time_series_covid19_confirmed_us_1$Province_State),]
head(time_series_covid19_confirmed_maryland_1)
#### another way to filter the cases by state is to use dplyr's filter function
library(dplyr)
time_series_covid19_confirmed_florida_1 <- filter(time_series_covid19_confirmed_us_1, Province_State=="Florida")
head(time_series_covid19_confirmed_florida_1)
```

```r
## Section 6d - creating new column and obtaining row sum
#### here we will create a new column to account for the total cases reported by each county from 10/30/2021 to 11/3/2021
#### we will use the apply command to obtain the sum of case counts for those dates
time_series_covid19_confirmed_maryland_1$county_total <- apply(time_series_covid19_confirmed_maryland_1[,c(3:7)],1,sum)
head(time_series_covid19_confirmed_maryland_1)
```

```r
## Section 6e - sorting
#### sort by county total case counts from largest to smallest using the order command
time_series_covid19_confirmed_maryland_1_sort1 <- time_series_covid19_confirmed_maryland_1[order(-time_series_covid19_confirmed_maryland_1$county_total),]
head(time_series_covid19_confirmed_maryland_1_sort1)
write.csv(time_series_covid19_confirmed_maryland_1_sort1, file="time_series_covid19_confirmed_maryland_1_sort1.csv", row.names=FALSE)
```

```r
## Section 6f - calculate column sum and the percent contribution of each column element to column total 
time_series_covid19_confirmed_maryland_2 <- time_series_covid19_confirmed_maryland_1_sort1
#### using apply to calculate total covid cases reported in maryland for each day from 10/30/2021 to 11/3/2021
time_series_covid19_confirmed_maryland_2_day_total <- apply(time_series_covid19_confirmed_maryland_2[,c(3:7)],2,sum)
#### set up loop to calculate on each day, the percent contribution to the total case counts for each county
for (a3 in 1:5)
    {time_series_covid19_confirmed_maryland_2[,a3+2] <- time_series_covid19_confirmed_maryland_2[,a3+2]*100/time_series_covid19_confirmed_maryland_2_day_total[a3]}
time_series_covid19_confirmed_maryland_2$county_total <- NULL
head(time_series_covid19_confirmed_maryland_2)
```

```r
## Section 6g - rearranging data values into one column and grouping
#### here, we will move all of the Maryland covid case counts from 10/30/2021 to 11/3/2021 to one column
#### and group them by date in which the cases were reported
#### to do this, we will use the melt function in the package reshape2
#### the rationale for doing this is that it will help us with producing a stacked bar plot using ggplot2
library(reshape2)
time_series_covid19_confirmed_maryland_3 <- melt(time_series_covid19_confirmed_maryland_2)
head(time_series_covid19_confirmed_maryland_3)
```

```r
## Section 6h - write a data frame to csv file to working directory
write.csv(time_series_covid19_confirmed_maryland_1_sort1, file="time_series_covid19_confirmed_maryland_1_sort1.csv")
write.csv(time_series_covid19_confirmed_maryland_2, file="time_series_covid19_confirmed_maryland_percent_total.csv", row.names=FALSE)
```

```r
## Section 8a - x, y linear scatter plot
## the parameter cex specified in the plot command controls sizes of several plot features
#### cex.lab alters the size of axis labels
#### cex.axis alters the size of tick mark labels
#### cex.main controls the size of the plot title
#### the default size for cex is 1
#### setting cex=1.5 indicates a 50% increase in size, whereas setting cex=0.5 indicates a 50% decrease in size
#### so really the sizes set by cex is all relative to the default of 1
#### see https://www.statmethods.net/advgraphs/parameters.html for cex
## las dictates the orientation of the tick mark lables
#### las=0 - default
#### las=1 - always horizontal
#### las=2 - always perpendicular
#### las=3 - always vertical
#### see http://rfunction.com/archives/1302 for las
x <- c(1,2,3,4) 
y <- c(2,4,6,8)
plot(x,y, cex.lab=2, cex.axis=1.25,cex.main=2,main="example scatter plot",
las=1)
```

```r
## Section 8b - changing plot symbols
#### the pch parameter in plot allows us to choose the symbol shape (for instance, open circle or solid square)
#### see http://www.sthda.com/english/wiki/r-plot-pch-symbols-the-different-point-shapes-available-in-r for pch
x <- c(1,2,3,4)
y <- c(2,4,6,8)
plot(x,y,cex.lab=2,cex.axis=1.25,pch=15,cex=2,las=1)
```

```r
## Section 8c - plotting histograms using the hist function
#### here we will create a vector points_scored with points scored for each game by our favorite basketball team
#### breaks in the hist function specifies the bins
#### here we want the breaks or bins to go from 60 to 140 incrementing by 20 and we set this using the seq command
#### in the seq command, the first number correspond to start, the second number correspond to end, and the third number dictates the increments that we will use to go from start to end
#### ylim allows us to specify a vector containing the limits of the y axis
gene_counts <- c(65,65,70,75,80,100,90,110,120,130,75,60,65,66) 
hist(gene_counts, breaks=seq(60,140,20), ylim=c(0,10), xlab="gene_counts", col="green", cex.lab=1.5, cex.axis=1.5, cex.main=1.5, main="gene count distribution")
```

```r
## Section 8d - generating two plots in one plot window
## to generate two plots in one plot window, we can use par(mfrow)
## here, we will plot high temperatures for each day of the week as a bar plot and pie chart
days <- c("sunday", "monday", "tuesday", "wednesday", "thursday", "friday", "saturday")
temperature <- c(75, 80, 79, 85, 81, 83, 82)
## mfrow=c(1,2) creates a plot window with 1 row and 2 columns, we will have the bar plot and pie chart side by side
par(mfrow=c(1,2)) 
barplot(temperature, main="high temperatures for the week", ylab="high temperature", cex.axis=1.5, cex.lab=1.5, names=days, cex.name=1.25, las=2)
pie(temperature, main="high temperatures for the week", radius=0.75, labels=days)
## alternatively, we can use mfcol=c(2,1) and this will draw the bar plot on top of the pie chart
par(mfcol=c(2,1)) 
barplot(temperature, main="high temperatures for the week", ylab="high temperature", cex.axis=1.5, cex.lab=1.5, names=days, cex.name=1.25, las=2)
pie(temperature, main="high temperatures for the week", radius=1, labels=days)
```

```r
## Section 8e - bar plot of Maryland covid cases
par(mar=c(10,7.5,3,3))  
#### setting scipen tells R that for numbers up to 7 digits, do not use scientific notation for the y axis
options(scipen=7)
barplot(time_series_covid19_confirmed_maryland_1_sort1$county_total, names=time_series_covid19_confirmed_maryland_1_sort1$County, las=2, ylim=c(0,500000),
        main="total covid cases (10/30/2021 - 11/3/2021) in maryland by county")
#### the line parameter determines the location of x and y axis labels        
title(ylab="number of cases", xlab="county", line=6)
#### alternatively, we can set mgp parameters in par to tell R where to put the x and y axis labels
par(mar=c(10,7.5,3,3), mgp=c(6,0.75,0))
options(scipen=7)
barplot(time_series_covid19_confirmed_maryland_1_sort1$county_total, names=time_series_covid19_confirmed_maryland_1_sort1$County, las=2, ylim=c(0,500000),
        main="total covid cases (10/30/2021 - 11/3/2021) in maryland by county", xlab="county", ylab="number of covid cases") 
```

```r
## Section 8f - creating stacked bar plot using the barplot command
#### here we will generate a stacked bar plot for county covid case percentage contributions from 10/30/2021 - 11/3/2021
library(RColorBrewer)
## see https://r-charts.com/part-whole/stacked-bar-graph/ for more on stacked bar plots
## setting border to white allows us to create a "white break" between the touch down bar and field goal bar
## the legend parameter allows us to specify the legend of the stacked bar plot
## args.legend dictates where the legend is located on the plot
coloring_1 <- brewer.pal(9, "Set1")
coloring_2 <- brewer.pal(8, "Set2")
coloring_3 <- brewer.pal(9, "Set3")
par(mar=c(4,5,3,10))
barplot(as.matrix(time_series_covid19_confirmed_maryland_2[,3:7]), border="white", col=c(coloring_1,coloring_2,coloring_3),
las=1, ylab="counts-%total", xlab="date", cex.axis=1.5, names=c("10/30","10/31","11/1","11/2","11/3"), legend=as.vector(time_series_covid19_confirmed_maryland_2$County),
args.legend=list(x="right",inset=c(-0.4,0)))
```

```r
## Section 8g - bar plot using ggplot2
#### note we will have to relevel the counties so that it plots from largest total case count to smallest total case count
#### within the aes parameter, we specify the y and x values - here we want to plot case count (y) by county (x)
#### geom_bar tells R that we want a bar plot (other plots such as box plot can be specified using geom_boxplot)
#### we set stat="identity" to tell ggplot2 that we provided the y values for the plot (see https://stackoverflow.com/questions/59008974/why-is-stat-identity-necessary-in-geom-bar-in-ggplot)
#### themes in ggplot2 allows us to set many plot features including plot background color and text size of axis elements
#### see https://ggplot2.tidyverse.org/reference/theme.html for more on themes
#### in this plot, in themes, axis.text.x adjusts the text size and orientation of the x axis tick mark labels
#### axis.title adjusts the text size of the x and y axis labels
#### panel.grid.minor and panel.grid.major allows us to turn on or off the grid lines in the plot
#### axis.line.x and axis.line.y allows us to modify the line shown in the x an y axes
#### panel.background allows us to adjust the plot background color
library(ggplot2)
par(mar=c(10,7.5,3,3))
time_series_covid19_confirmed_maryland_1_sort1[,1] <- 
factor(time_series_covid19_confirmed_maryland_1_sort1[,1], 
levels=time_series_covid19_confirmed_maryland_1_sort1[,1][order(-time_series_covid19_confirmed_maryland_1_sort1[,8])])
maryland_covid_plot1 <- ggplot(time_series_covid19_confirmed_maryland_1_sort1, aes(y=county_total, x=County))+geom_bar(stat="identity")
maryland_covid_plot1 <- maryland_covid_plot1+theme(axis.text.x=element_text(size=10, angle=90, hjust=0.99), axis.title=element_text(size=12,face="bold"), panel.grid.minor=element_blank(), panel.grid.major=element_blank(),
axis.line.x=element_line(color="black"), axis.line.y=element_line(color="black"), panel.background=element_rect(fill="white"))
maryland_covid_plot1 <- maryland_covid_plot1+labs(title="covid cases (10/30/2021-11/3/2021) in maryland by county")
plot(maryland_covid_plot1)
```
